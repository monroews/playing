```
from aguaclara.core.units import unit_registry as u
from aguaclara.core import physchem as pc
from aguaclara.core import utility as ut
from aguaclara.core import pipes as pipes
import numpy as np
import matplotlib.pyplot as plt
```

```
# volts to cm (cm/volt) for 100 psi
P = 100 * u.psi
density = pc.density_water(22*u.degC)

x = ((P/(density*u.gravity))/(100*u.mV)).to(u.cm/u.V)
print('The conversion from volts to cm for a',P,'sensor is',x)
# 1mV=1psi
```
