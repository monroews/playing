```
""" importing """
from aguaclara.core.units import unit_registry as u
import aguaclara as ac
import matplotlib.pyplot as plt
import numpy as np
# https://en.wikipedia.org/wiki/Fuel_economy_in_aircraft

Boeing737Fuel = 2.43 * u.L/(100 * u.km)
EnginContext_distance = 2* 2000 * u.mile
EnginContext_N = 20/u.year
Fuel = (Boeing737Fuel*EnginContext_distance*EnginContext_N).to(u.L/u.year)
print(Fuel)
# https://www.eia.gov/environment/emissions/co2_vol_mass.php
CO2_Per_mass_JetFuel = 9.57* u.kg /u. gallon
Engin_Context_CO2 = (CO2_Per_mass_JetFuel * Fuel).to(u.kg/u.year)
print('The CO2 produced by one trip is',Engin_Context_CO2)
Mechanized_water_energy = 100 * u.J/u.L
#g of CO2 per GJ of electricity
# https://en.wikipedia.org/wiki/Fossil_fuel_power_station
Bunker_electricity = 77400 * u.g/u.GJ
print('Grams of CO2 per GJ of electricity produced using bunker fuel',Bunker_electricity)
Water_CO2 = (Mechanized_water_energy * Bunker_electricity).to(u.g/u.L)
print('Mechanized package water treatment plants produce this many mg of CO2 per liter of water', Water_CO2.to(u.mg/u.L))
#How many liters of water do we need to produce?
Water_Produced = (Engin_Context_CO2 / Water_CO2).to (u.L/u.s)
print('The amount of water AguaClara needs to produce to offset the air travel', Water_Produced)
```
